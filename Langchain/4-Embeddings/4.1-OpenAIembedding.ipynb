```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model = "text-embedding-3-large")
embeddings
```

```python
text = "This is a tutorial on OPENAI embeddings"
query_result = embeddings.embed_query(text)
query_result
```

```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
embeddings_1024
```

```python
text = "This is a tutorial on OPENAI embeddings"
query_result = embeddings_1024.embed_query(text)
query_result
```

```python
## Text Loader
from langchain_community.document_loaders import TextLoader

loader = TextLoader('speech.txt')
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 50)
final_docs = text_splitter.split_documents(docs)

final_docs
```

```python
## Vector Embeddig and Vector StoreDB
from langchain_community.vectorstores import Chroma

## Convert text into vectors using openAi embeddings and store in ChromaDB
db = Chroma.from_documents(final_docs, embeddings_1024)

db
```

```python
## retrived the similar docs from the vector DB
query = "It is a distressing and oppressive duty"
retrived_res = db.similarity_search(query)
print(retrived_res)

```
