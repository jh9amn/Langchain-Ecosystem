Ollama

Ollama supports embedding models, making it possible to build retrival augumented generation (RAG) applications that combine text prompts with existing documents or other data.

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")    ## by default it uses "llama2" model
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_query(
    ["Alpha is the first letter of the Greek alphabet.",
     "Beta is the second letter of the Greek alphabet."]
)
```

```python
r1
```

```python
r1[0]
```

```python
len(r1)
```

```python
embeddings.embed_query("What is the second letter of the Greek alphabet?")
```
