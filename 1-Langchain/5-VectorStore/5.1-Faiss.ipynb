Facebook AI Similarity Search (FAISS) is a library for efficient similarity search and clustering of dense vectors. It is particularly useful for applications such as information retrieval, recommendation systems, and machine learning, where you need to find similar items quickly from large datasets.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size = 1000, chunk_overlap = 0)
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embeddings = OllamaEmbeddings(model="mxbai-embed-large:latest")
db = FAISS.from_documents(docs, embeddings)
db
```

```python
## Querying the vector store
query = "How does the speaker describe the desired outcome of the war?"
query_res = db.similarity_search(query)
query_res[0].page_content
```

```python
query_res
```

#### As a Retriever
We can also comvert the vector store into a Retriver class. This allows us to use it more easily with Langchain chains and LLMs, which largerly work with retrivers.

```python
retriever = db.as_retriever()
retriever.invoke(query)
docs[0].page_content
```

#### Similarity Seach with score

There are sime FAISS specific methods. One of them is similarity_Search_with_score, which allows you to return not only the documents but also the distance socre of the query to them. The returned distance socre is L2 disctance. Therefore, a lower score is better.

```python
docs_and_score =db.similarity_search_with_score(query)
docs_and_score
```

```python
embedding_vector = embeddings.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```

```python
### Saving and loading the vector store
db.save_local("faiss_index")
```

```python
new_db = FAISS.load_local("faiss_index", embeddings, allow_dangerous_deserialization=True)
docs = new_db.similarity_search(query)
```
